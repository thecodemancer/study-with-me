```
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os

%matplotlib inline

sns.set()
x = np.array( range(0,20) )
y = np.random.choice(x, size = 20)
type(x)
```

# Line Chart

```
plt.figure(figsize=(10,5))
ax = sns.lineplot(x=x,y=y)
ax.set_xlabel(xlabel="X Axis", fontsize=10)
ax.set_ylabel(ylabel="Y Axis", fontsize=10)
ax.set_title(label="The title goes here", fontsize=15)
```

# Bar charts

```
from collections import Counter
#x= df['variety']
counter = Counter(x)
height = list(counter.values())
labels = list(counter.keys())
plt.figure(figsize=(12,8))
ax=sns.barplot(labels, height)
ax.set_xlabel("Species")
ax.set_ylabel("Frequency")
ax.set_title("Species Distribution")
```

```
ax.figure.savefig("Iris bart chart - Frequency Vs Species.png")
```

```
plt.figure(figsize=(12,8))
ax=sns.barplot(x="Species", y="SepalLengthCm", data=df, estimator=sum)
ax.set_xlabel("Species")
ax.set_ylabel("Frequency")
ax.set_title("Species distribution")
```

```
plt.figure(figsize=(12,8))
ax=sns.barplot(x="Species", y="SepalLengthCm", data=df, estimator=np.std)
ax.set_xlabel("Species")
ax.set_ylabel("Frequency")
ax.set_title("Species distribution")
```
